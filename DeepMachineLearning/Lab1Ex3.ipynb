```python
import torch
t1 = torch.rand(size=(3,3,3,3))
t2 = torch.rand(size=(3,3,3,3))
```

```python
t1.dtype
```

```python
t1.cuda()
t2.cuda()
```

```python
t3 = torch.sum(t1)
t3
```

```python
t4 = torch.add(t1, t2)
t4
```

```python
t5 = t1 + t2
t5    #same of add
```

```python
t6 = t1 @ t2
t6    #same of matmul
```

```python
t9 = torch.matmul(t1, t2)
t9
```

```python
t7 = t1 * t2
t7    #same of mul
```

```python
t8 = torch.mul(t1, t2)
t8
```

```python
t10 = torch.cat((t1, t2), axis=3)
t10.size()
```
